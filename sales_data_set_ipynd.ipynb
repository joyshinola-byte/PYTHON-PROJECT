```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv(file_path)
```

```python
sales_array = data["Sales_Amount"].to_numpy()
quantity_array = data["Quantity_Sold"].to_numpy()
data['Profit'] = data['Sales_Amount'] - (data['Quantity_Sold'] * data['Unit_Cost'])
profit_array = data["Profit"].to_numpy()

print("Sales Array:", sales_array[:5])
```

```python
quantity = data['Quantity_Sold']
unit_price = data['Unit_Price']
revenue = quantity * unit_price
total_sales = np.sum(revenue)
average_sales = np.mean(revenue)

print("\nTotal Sales:", total_sales)
print("Average Sales:", average_sales)
```

```python

print("\nFirst Record:", revenue[0])
print("First 3 Records:", revenue[:3])
print("Last 2 Records:", revenue[-2:])
```

```python
mean_sales = np.mean(revenue)
median_sales = np.median(revenue)
variance_sales = np.var(revenue)
std_deviation = np.std(revenue)

print("\nMean:", mean_sales)
print("Median:", median_sales)
print("Variance:", variance_sales)
print("Standard Deviation:", std_deviation)
```

```python
discount_price = unit_price * 0.9
discount_revenue = quantity * discount_price

print("\nRevenue after 10% Discount:\n", discount_revenue)
```

```python
performance = np.where(revenue > average_sales,
                       "High Sales",
                       "Low Sales")

print("\nPerformance Analysis:\n", performance)
```
